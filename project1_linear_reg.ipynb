```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn import linear_model
```

```python
path="C:/Users/James Bond/Downloads/canada_per_capita_income.csv"
df = pd.read_csv(path)
print(df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["year"]],df["per capita income (US$)"])
```

```python
reg.predict([[2030]])

```

```python
reg.coef_
```

```python
%matplotlib inline
plot.scatter(df.year,df["per capita income (US$)"],marker="+")
plot.ylabel("per capita income (US$)")
plot.xlabel("year")
```

```python
%matplotlib inline
plot.scatter(df["year"],df["per capita income (US$)"],marker = "+")
plot.xlabel("year")
plot.ylabel("per capita income (US$)")
plot.plot(df.year,reg.predict(df[["year"]]))

```
